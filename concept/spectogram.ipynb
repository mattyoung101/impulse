```python
import numpy as np
import matplotlib.pyplot as plt
import soundfile as sf

PATH = "sawtooth.flac"

audio, samplerate = sf.read(PATH)
print(f"Sample rate: {samplerate}")

plt.figure(figsize=(10, 8), dpi=80)
plt.set_cmap("inferno")
plt.specgram(x=audio, Fs=samplerate)
plt.colorbar()
plt.show()
```
